```python
city_for_ml2 = city_for_ml[['state_abbr', 'city_name', 'metric_name', 'est']].copy()
city_for_ml2.head()
```

```python
city_for_ml2['uniq_name'] = city_for_ml2['state_abbr'] + '_' + city_for_ml2['city_name']
```

```python
city_for_ml2.drop(['state_abbr', 'city_name'], axis=1, inplace=True)
```

```python
city_for_ml2.set_index('uniq_name', inplace=True)
```

```python
city_for_ml2.head()
```

```python
city_for_ml2['metric_name'] = city_for_ml2['metric_name'].str.lower().str.replace(' ', "_")
```

```python
city_pivot = city_for_ml2.pivot(columns='metric_name', values='est')
```

```python
city_pivot.head()
```

```python
city_pivot.info()
```

```python
# low_birthweight, prenatal_care, and teen_births are missing < 200 values - remove columns
city_pivot.drop(['low_birthweight', 'prenatal_care', 'teen_births'], axis=1, inplace=True)
```

```python
city_pivot.describe()
```

```python
opioid_corr = pd.DataFrame(city_pivot.corr()['opioid_overdose_deaths']).reset_index()
```

```python
opioid_corr_kendall = pd.DataFrame(city_pivot.corr(method='kendall'))['opioid_overdose_deaths'].reset_index()
opioid_corr_spear = pd.DataFrame(city_pivot.corr(method='spearman'))['opioid_overdose_deaths'].reset_index()
```

```python
opioid_corr.columns = ['metric_name', 'opioid_corr_pear']
opioid_corr_kendall.columns = ['metric_name', 'opioid_corr_kend']
opioid_corr_spear.columns = ['metric_name', 'opioid_corr_spear']
#opioid_corr.merge(opioid_corr_kendall)
```

```python
# opioid data has a non-normal distribution - use spearman correlation
opioid_corr_spear.head()
```

```python
opioid_corr_spear.sort_values('opioid_corr_spear', inplace=True)
```

```python
sns.set(font_scale=1.5)
plt.figure(figsize=(10,16))
sns.barplot(y='metric_name', x='opioid_corr_spear', data=opioid_corr_spear, orient="h")
plt.xlabel('Correlation Coeff (Spearman)')
plt.ylabel('Metric Name')
plt.title('Correlation between opioid overdose deaths and all metrics\n(500 Cities; Total Population)')
plt.show()
```
